# Electro-Optical

- visible: 0.75 - 0.3$\mu$m
- IR: 1 - 100$\mu$m

```python
import numpy as np
from math import pi, atan2
from matplotlib import pyplot as plt
```

```python
max_size = 4 # meters
```

```python
fpa = (480,640,)
pixel_res = max_size / fpa[0]
print(f"Pixel res: {pixel_res * 1000:.0f} mm/px")
```

```python
fovs = []
ranges = list(range(100,1000))
for m_size in range(1,5):
    f = []
    for r in ranges:
        f.append(2*atan2(m_size/2, r)*180/pi)
    fovs.append(f)
```

```python
plt.figure(figsize=(20,10))
for s,y in enumerate(fovs):
    plt.plot(ranges, y, label=f"{2*(s+1)} m")
plt.xlabel("range [m]")
plt.ylabel("FOV [deg]")
plt.legend()
plt.grid(True);
```
