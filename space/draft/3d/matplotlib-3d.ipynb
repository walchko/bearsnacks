```python
def axes(r, ax=None, color=None):
    if not isinstance(r, np.ndarray):
        r = np.array(r)
        
    x = r[:,0]
    y = r[:,1]
    z = r[:,2]
    
    # https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
    ls = "solid"
    
    if ax is None:
        # not sure how to fix this, lots of open figure stacking up in 
        # background ... not this helps
        # plt.close('all')
        
        # ax = plt.figure().add_subplot(projection='3d')
        fig = plt.figure()
        rect = 0,0,1,1
        ax = fig.add_axes(rect, projection='3d')
        ax.text(x[0], x[1], x[2], "x")
        ax.text(y[0], y[1], y[2], "y")
        ax.text(z[0], z[1], z[2], "z")

        ax.set_xlim([-1.5, 1.5])
        ax.set_ylim([-1.5, 1.5])
        ax.set_zlim([-1.5, 1.5])
        
        # clear tick marks, don't need them
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.set_zticklabels([])
        
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.set_zlabel("z")
        
        # ax.set_aspect('equal')
        # ax.set_axis_off()
        colors = ["r","g","b"]
        ls = "dashdot"
        # fig.close()
        # help(fig)
    
    elif color is None:
        color = "#aaaaaa"
        colors = [color, color, color]
    else:
        colors = [color, color, color]
    
    for a,c in zip([x,y,z],colors):
        ax.quiver(
            [0],[0],[0],
            a[0],a[1],a[2],
            color=[c],
            # length=0.05,
            # normalize=True,
            linestyle=ls,
        )
        
    # for i in range(1,plt.gcf().number+1):
    #     plt.close(i)
    # plt.close('all')

    return ax
    
q = Quaternion.from_euler(0,0,0,degrees=True)
r = q.to_rot()

p = axes(r)
p.set_title("hello")

q = Quaternion.from_euler(45,-45,0,degrees=True)
r = q.to_rot()

p = axes(r,p,color="c")

q = Quaternion.from_euler(45,-45,45,degrees=True)
r = q.to_rot()

# plt.subplot(1,2,1)
p = axes(r,p,color="#aaaaaa")
# p.add_subplot(1,2,1)
# p.show()
```

```python
def test(f, x, sp=None):
    if sp is not None:
        ax = f.add_subplot(*sp)
    else:
        ax = f.gca()
    ax.plot(x)
    ax.grid(True)
    return f

f = plt.figure(figsize=(10,5))
f = test(f, range(10), sp=(1,2,1))
f = test(f, range(5), sp=(1,2,2))
f.gca().set_title("second")
f.get_axes()[0].set_title("first title")
f.get_axes()[0].set_xlabel("first x");
f.get_axes()[1].set_xlabel("second x");
```
