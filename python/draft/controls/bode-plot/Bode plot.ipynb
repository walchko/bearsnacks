# Simple Low Pass Filter Bode Plot

```python
%matplotlib inline
```

```python
from __future__ import print_function
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
from math import pi, sqrt, atan2
```

```python
x=[]
y=[]
y2=[]
R = 10
C = 1/100
for i in range(0,100000):
    x.append(i)
    y.append(1/sqrt((i*R*C)**2+1))
    y2.append(180-atan2(i*R*C,1)*180/pi)
#     y2.append(-atan2(-i*R*C,-1)*180/pi)
```

```python
plt.subplot(2,1,1)
plt.semilogx(x,y)
plt.grid(True)
plt.ylabel('$|H(j\\omega)|$')

plt.subplot(2,1,2)
plt.semilogx(x,y2)
plt.grid(True)
plt.ylabel('$\\angle H(j\\omega)$')
plt.xlabel('frequency ($\\omega$)');
plt.savefig('response.png')
```
